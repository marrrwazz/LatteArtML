```python
from skimage import io
from skimage.transform import resize
from skimage.color import rgb2grey
from os import listdir
import numpy as np
```

Here we load in our presorted data as images labeled good or bad.

```python
goodFiles = listdir("GoodArt")
badFiles = listdir("BadArt")
```

Below we load in each of our images, convert the image to gray scale, resize to a 256 by 256 pixel, and flatten into row vector. We keep a counter to keep track of how many images were processed.

```python
goodArt_Transformed = []
i = 1
for file in goodFiles:
    print(i)
    #take each image, convert to gray scale, resize all images to same size
    #flatten so that on row is one image
    if ".jpg" in file:
        i = i +1
        file = "GoodArt/"+file
        original = io.imread(file)
        original = rgb2grey(original)
        good_tmp = resize(original,(256,256))
        good_tmp = good_tmp.flatten()
        goodArt_Transformed.append(good_tmp)
```

```python
badArt_Transformed = []
i = 1
for file in badFiles:
    if ".jpg" in file:
        print(i)
        i = i +1
        file = "BadArt/"+file
        original = io.imread(file)
        original = rgb2grey(original)
        bad_tmp = resize(original,(256,256))
        bad_tmp = bad_tmp.flatten()
        badArt_Transformed.append(bad_tmp)
```

In the three cells below we create a categorical variable variable for the class label and concatenate our data to have a target matrix and an input matrix.

```python
goodLabels = np.repeat("Good",len(goodArt_Transformed))
badLabels = np.repeat("Bad",len(badArt_Transformed))
```

```python
goodArt = np.array(goodArt_Transformed)
goodLabels = np.array(goodLabels)
good = np.append(goodArt,goodLabels[:,None],axis=1)

badArt = np.array(badArt_Transformed)
badLabels = np.array(badLabels)
bad = np.append(badArt,badLabels[:,None],axis=1)

latte_pics_transformed = np.append(good,bad)
```

```python
latte_data = np.concatenate((goodArt,badArt),axis=0)
latte_labels = np.concatenate((goodLabels,badLabels),axis=0)
```

We use the standard train_test_split function to generate a random test set and training set, with 80% of our images going towards training.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(latte_data,latte_labels,test_size=0.2)
```

The first model we will try is a Random Forest model, with the all of the default settings. We will also use the GridSearchCV method to find optimal parameters.

```python
from sklearn.ensemble import RandomForestClassifier

rForest = RandomForestClassifier(warm_start=True)
rForest.fit(X_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators':(10,20,50,100,200),'max_depth':(2,10,20,50,100,None),'class_weight':("balanced","balanced_subsample",None)}

rForestCV = RandomForestClassifier(verbose=True)
clf = GridSearchCV(rForestCV,parameters)
clf.fit(X_train,y_train)
```

```python
bestRF = clf.best_estimator_
bestRF.fit(X_train,y_train)
bestRF.score(X_test,y_test)
```

Our best Random Forest model had a accuracy score of about 54.4%. Not much better than flipping a coin. Below we try a neural network with the standard out of the box settings and obtain an accuracy of about 58% on our test set. We use the same GridSearchCV method to find an optimal set of parameters, only changing the solver setting. This results in an accuracy of 54.8% on our test set.

```python
from sklearn.neural_network import MLPClassifier

nnet = MLPClassifier()
nnet.fit(X_train,y_train)
nnet.score(X_test,y_test)
```

```python
parameters = {'solver':('lbfgs','sgd','adam')}
nnetCV = MLPClassifier(verbose=True)
clfNN = GridSearchCV(nnetCV,parameters)
clfNN.fit(X_train,y_train)
```

```python
bestNN = clfNN.best_estimator_
bestNN.fit(X_train,y_train)
bestNN.score(X_test,y_test)
```

The next model we try is the KNN model and we try again to do a search to optimize the number of neighbors we use. As it turns out the GridSearchCV returns a value of 5 for K, which produces an accuracy of 57.4% on our test set.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
parameters = {'n_neighbors':(1,3,5,7,9)}
knnCV = KNeighborsClassifier()
clfCV = GridSearchCV(knnCV,parameters)
clfCV.fit(X_train,y_train)
```

```python
bestKNN = clfCV.best_estimator_
bestKNN.fit(X_train,y_train)
bestKNN.score(X_test,y_test)
```

We try to run the gaussian and multinomial naive-bayes models on our data and return with similar results.

```python
from sklearn.naive_bayes import MultinomialNB, GaussianNB

nb = MultinomialNB()
nb.fit(X_train,y_train)
print("Multinomial Naive-Bayes")
print("Score on test: ",nb.score(X_test,y_test))
print("Score on train: ",nb.score(X_train,y_train))

gnb = GaussianNB()
gnb.fit(X_train,y_train)
print("Gaussian Naive-Bayes")
print("Score on test: ",gnb.score(X_test,y_test))
print("Score on train: ",gnb.score(X_train,y_train))
```

Nearly identical results are obtained with an AdaBoost model. These models take quite a long time to train so we chose 10, 50, and 100 as values for the number of estimators. As it turns out, 10 estimators produces the best accuracy, but again only 54.8%. 

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier()
ada.fit(X_train,y_train)
print("Ada Boost Classifier")
print("Score on test: ",ada.score(X_test,y_test))
print("Score on train: ",ada.score(X_train,y_train))
```

```python
ada10 = AdaBoostClassifier(n_estimators=10)
ada10.fit(X_train,y_train)
ada10.score(X_test,y_test)
```

```python
ada100 = AdaBoostClassifier(n_estimators=100)
ada100.fit(X_train,y_train)
ada100.score(X_test,y_test)
```
